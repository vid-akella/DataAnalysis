```python
#import packages
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Analysis of Lipid Data

The dataset contains health and lifestyle data from 45 individuals, both male and female adults, in an age range of 20-40 years. A sample of the data table is below:

```python
#read in data
lipid = pd.read_csv("/Users/vidya/Library/CloudStorage/OneDrive-Personal/GitCode/DataAnalysis/lipid2.csv",  sep=",")
lipid.head()
```

Data from males and females were separated since they are expected to have different distributions and health. Plotting frequencies for Weight ranges in Males and Females.

```python
#plot frequencies for weight, cholesterol and LDL categories
plt.hist([lipid['Weight'][lipid["Sex"] == "M"],lipid['Weight'][lipid["Sex"] == "F"]], bins=10, color=[ 'blue', 'purple'], label=['Male', 'Female'])
plt.legend()
plt.show()

```

Distribution of Cholesterol measurements in dataset.

```python
#cholesterol
plt.hist([lipid['Cholesterol'][lipid["Sex"] == "M"],lipid['Cholesterol'][lipid["Sex"] == "F"]], bins=10, color=[ 'blue', 'purple'], label=['Male', 'Female'])
plt.legend()
plt.show()
```

Distribution of LDL values in the dataset

```python
#LDL
plt.hist([lipid['LDL'][lipid["Sex"] == "M"],lipid['LDL'][lipid["Sex"] == "F"]], bins=10, color=[ 'blue', 'purple'], label=['Male', 'Female'])
plt.show()
```

Examining the relationship between weight and Cholesterol. Is cholesterol higher in individuals that weigh more?

```python
sns.lmplot(x="Weight", y="Cholesterol", data=lipid, col="Sex", hue="Sex", palette=['purple', 'blue'])
plt.show()
```

Examining relationship between Age and LDL

```python
sns.relplot(x=lipid["Age"], y=lipid["LDL"],  kind='scatter', hue=lipid["Sex"], palette=['purple', 'blue'])
plt.show()
```

Examining change in LDL and amount of exercise 

```python
sns.relplot(x=lipid["Exercise"], y=(lipid["Weight"]-lipid["Weight3"]),  kind='scatter', hue=lipid["Sex"], palette=['purple', 'blue'])
plt.show()
```
